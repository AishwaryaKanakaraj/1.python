#Manger needs (if different input age is given 23,45,67,34,89 it should tell output with the suitable condition)
age<18-->children
age<35-->adult
age<59-->citizen
age>59-->senior citizen

```python
lists=[23,45,67,34,89]
for age in lists:
    if(age<18):
        print("children")
    elif(age<35):
        print("adult")
    elif(age<59):
        print("citizen")
    else:
        print("senior citizen")
    
```
