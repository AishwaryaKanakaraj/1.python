```python
# OddEven # Create a function that checks whether the given number is Odd or Even # WITHOUT RETURN
```

```python
def OddEven():
    num=int(input("Enter the number"))
    if((num%2)==1):
        print(num,"is Odd number")
    else:
        print(num,"is Even number")
    
```

```python
# Create a function that tells elegibility of marriage for male and female according to thei
r age limit like 21 for male and 18 for female
Your Gender:Male
Your Age:20
NOT ELIGIBLE
```

```python
def ElegiblityForMarriage():
    gender= input("Your Gender:")
    age=int(input("Your age:"))
    if(gender=="Male" and age >=21):
        print("ELIGIBLE")
    elif(gender=="Female" and age >=18):
        print("ELIGIBLE")
    else:
        print("NOT ELIGIBLE")
    return
```

```python
ElegiblityForMarriage()
```
